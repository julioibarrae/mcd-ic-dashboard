```python
#Declaramos una funcion para descargar los archivos por símbolo

def descargarArchivo(symbol): 
    #Calculamos fecha y obtenemos el nombre del archivo
    fechainicio = datetime.datetime(2019, 1, 1)
    if(datetime.datetime.utcnow().hour >= 20):
        fechafinal = datetime.datetime.utcnow()
    else:
        fechafinal = datetime.datetime.utcnow() + datetime.timedelta(days=-1)
    fechafinal = fechafinal.replace(hour=0, minute=0, second=0, microsecond=0)
    archivo_nombre = "data/" +  symbol + "_" +  fechainicio.strftime('%Y%m%d') + "_"+ fechafinal.strftime('%Y%m%d') + ".csv"
       
    #Convertimos la fecha a timestamp para la consulta.
    fechainicio =  fechainicio.replace(tzinfo=datetime.timezone.utc).timestamp()
    fechafinal = (fechafinal + datetime.timedelta(days=1)).replace(tzinfo=datetime.timezone.utc).timestamp()
    url= "https://query1.finance.yahoo.com/v7/finance/download/%5E" + symbol + "?period1=" + str(int(fechainicio)) + "&period2=" + str(int(fechafinal)) + "&interval=1d&events=history&includeAdjustedClose=true"
       
    if not os.path.exists(archivo_nombre):
        urllib.request.urlretrieve(url, archivo_nombre)
```

```python
indices = ["DJI"      #Dow Jone Industrial Average
           , "RUA"    #Russell 3000 Index
           , "RUT"    #Russell 2000 Index
           , "GSPC"   #S&P 500
           , "IXIC"   #Nasdaq Composite
           , "VIX"]   #Expectation of volatily
for i in indices:
    descargarArchivo(i)
```
